The problem can be stated in this way; in n-dimension, how many Boolean function are there such that the convex hulls of two disjoint sets will not intersect? Here single-layer-perceptron has been used to identify the number of Boolean function with 2, 3 & 4 input. Due to the lack of resource for computation, it was not possible to generate data programmatically.

#### Now question is there; can we formulate the number of linearly separable  Boolean functions for n- dimension?

#### For 2 dim, we get 16 function (14 are linearly separable).
#### For 3 dim, we get 256 function (104 are linearly separable).
#### For 4 dim, we get 65536 function (1882 are linearly separable).

#### The all above data has been found programmatically. But no formulation is done so far.

```python
class Perceptron:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.activation_func = self._unit_step_func
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # init parameters
        self.weights = np.zeros(n_features)
        self.bias = 0

        y_ = np.array([1 if i > 0 else 0 for i in y])

        for _ in range(self.n_iters):

            for idx, x_i in enumerate(X):

                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation_func(linear_output)

                # Perceptron update rule
                update = self.lr * (y_[idx] - y_predicted)

                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self.activation_func(linear_output)
        return y_predicted

    def _unit_step_func(self, x):
        return np.where(x >= 0, 1, 0)
```

```python
def accuracy(y_true, y_pred):
        accuracy = np.sum(y_true == y_pred) / len(y_true)
        return accuracy

def illust(j, df_final,  acc, df_off, graph_off, show_non_lin):
    Status_var=""  
    if( acc<1):
      Status_var="*****(NON-LINEAR)*****"
    else:
      Status_var=""
    if (show_non_lin==1 and acc==1):
      df_off=0
      graph_off=0

    
    if (df_off==1):
     print("#", j+1," th Function --", Status_var, "\nPerceptron classification accuracy", acc ,"===")
     print(df_final,"\n================================================", Status_var)
    my_cmap = plot.get_cmap('hsv')
    if input_space==2 and graph_off==1:
      fig = px.scatter( df_final, x=0, y=1, color="Out", width=400, height=400)
      fig.update_traces(marker_size=10)
      fig.show()
    if input_space==3 and graph_off==1:
      seaborn.set_style("darkgrid")
      plot.figure(figsize=(6,5))
      axes = plot.axes(projection='3d')
      print(type(axes))
      colors = {1:'orange', 0:'blue'}
      color_ls = [colors[i] for i in df_final.Out.to_list()]
      axes.scatter3D(df_final[0], df_final[1], df_final[2], c = (color_ls), cmap = my_cmap, marker='o', s=100)
      axes.set_xlabel('x')
      axes.set_ylabel('y')
      axes.set_zlabel('z')
      plot.show()
      print("\n","\n","\n\n")


Non_Separate_Func=0

for j in range(df.shape[1]):
    output_df=pd.DataFrame({"Out":df[j].tolist()})
    df_final=pd.concat([input_df.T, output_df], axis=1)

    df_final_copy = df_final.copy()
    y=df_final.pop('Out')
    X_df=df_final
    X_arr = X_df.to_numpy()
    y_arr = y.to_numpy()

    p = Perceptron(learning_rate=0.01, n_iters=1000)
  
    p.fit(X_arr, y_arr)
    predictions = p.predict(X_arr)
    acc=accuracy(y_arr, predictions)
    #print("Perceptron classification accuracy", acc ,"\n","\n")
    if (acc<1):
      Non_Separate_Func=Non_Separate_Func+1

    illust(j,df_final_copy, acc, 0, 0,1)

print("Non Linearly Separable Boolean Function is:-", Non_Separate_Func)    
    
    
```
